```python
from matplotlib import pyplot as plt
import numpy as np
import statistics as st
```

```python
import random
print(random.randint(0,100))
```

```python
print(random.uniform(0,100))
```

```python
pop = []
for i in range(1000):
  pop.append(random.randint(0,100))
print(pop)
```

```python
fig,ax = plt.subplots(figsize=(10,7))
ax.hist(pop,bins =500)
print("Histogram of population shown below")
plt.show()
```

```python
print("Mean =",np.mean(pop))
print("Variance =",np.var(pop))
```

```python
n = int(input("Ënter sample size: "))
sample_data = random.sample(tuple(pop),n)
print(sample_data)
print("Mean of sample =",st.mean(sample_data))
print("Variance of sample =",st.variance(sample_data))
```

```python
n = int(input("Enter sample size: "))
m = int(input("Enter number of samples: "))
mean_sample=[]
for x in range(m):
  sample_data= random.sample(tuple(pop),n)
  mean = np.mean(sample_data)
  mean_sample.append(mean)
print("Mean of sample mean distribution:",np.mean(mean_sample))
print("Variance of sample mean distribution:",np.var(mean_sample))
```

```python
fig,ax = plt.subplots(figsize=(10,7))
ax.hist(mean_sample,bins =250)
print("Histogram of population shown below")
plt.show()
```

```python
#21
pop = []
for i in range(1000):
  pop.append(random.uniform(0,100))
fig,ax = plt.subplots(figsize = (10,7))
ax.hist(pop,bins = 500)
print("Histogram of population distribution given below: ")
plt.show()
print("Mean of population =",np.mean(pop))
print("Variance of population =",np.var(pop))
n = int(input("Enter sample size:"))
print("Predicted mean of sample mean distribution by Central Limit Theorem is:",np.mean(pop))
print("Predicted variance of sample mean distribution by Central Limit Theorem is:",(np.var(pop)/n))
m = int(input("Enter number of samples you want: "))
mean_sample = []
for x in range(m):
  sample_data = random.sample(tuple(pop),n)
  mean = st.mean(sample_data)
  mean_sample.append(mean)
print("Observed mean of Sampling mean distribution: ",np.mean(mean_sample))
print("Observed variance of Sampling Mean distribution",np.var(mean_sample))
print("Histogram of Sampling Mean Distribution shown below:")
fig,ax = plt.subplots(figsize = (10,7))
ax.hist(mean_sample,bins = 500)
plt.show()
```

```python
list_ones = list(np.zeros(395,np.int))
list_zeros = list(np.zeros(605,np.int))
pops=list_ones+list_zeros
fig, ax = plt.subplots(figsize = (10, 7))
ax.hist(pops, bins = 3)
print("Histogram of Population Distribution is shown below: ")
plt.show()
n=int(input("Enter sample size : "))
m=int(input("Enter no of samples : "))
sample_proportion=[]
for i in range (0,m):
    sample_data=random.sample(tuple(pops) , n)
    proportion=sum(sample_data)/n
    sample_proportion.append(proportion)
print("Histogram of sampling Distribution for Sample Proportion is shown below: ")
fig, ax = plt.subplots(figsize = (10, 7))
ax.hist(sample_proportion, bins = 500)
plt.show()

```
